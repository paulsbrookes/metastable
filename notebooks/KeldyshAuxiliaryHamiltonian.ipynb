<h1>Derivation of the Auxiliary Hamiltonian from the Keldysh Lagrangian</h1><p>This document presents a derivation of the auxiliary Hamiltonian for a driven–dissipative Kerr oscillator, starting from the Keldysh Lagrangian.</p><hr /><h2>1. Starting Point: The Keldysh Lagrangian</h2><p>The system is described by the Lagrangian written in terms of the forward (a_+) and backward (a_-) fields:</p><p>$$\begin{aligned}L &amp;= a_{+}^{<em>} \,i\partial_t a_{+} - a_{-}^{</em>}\, i\partial_t a_{-}+ i\varepsilon \Bigl( a_{+}^{<em>} - a_{+} - a_{-}^{</em>} + a_{-} \Bigr)\[1mm]&amp;\quad - \delta\Bigl(a_{+}^{<em>}a_{+} - a_{-}^{</em>}a_{-}\Bigr)- \chi\Bigl(a_{+}^{<em>2}a_{+}^2 - a_{-}^{</em>2}a_{-}^2\Bigr)\[1mm]&amp;\quad - i\kappa\Bigl( 2a_{+}a_{-}^{<em>} - a_{+}^{</em>}a_{+} - a_{-}^{<em>}a_{-}\Bigr).\end{aligned}$$The partition function is then given by$$\mathcal{Z} = \int \mathcal{D}a_{-}\,\mathcal{D}a_{-}^{</em>}\,\mathcal{D}a_{+}\,\mathcal{D}a_{+}^{<em>}\;e^{iS[a_+,a_+^</em>,a_-,a_-^*]}.$$</p><hr /><h2>2. Keldysh Rotation and Decomposition into Real Fields</h2><h3>2.1 Keldysh Rotation</h3><p>We introduce the <strong>classical</strong> and <strong>quantum</strong> fields by performing the Keldysh rotation:$$a_c = \frac{a_+ + a_-}{\sqrt{2}},\qquada_q = \frac{a_+ - a_-}{\sqrt{2}}.$$</p><h3>2.2 Decomposition into Real and Imaginary Parts</h3><p>Next, we express these complex fields in terms of their real components:$$a_c = \frac{x_c + i\,p_c}{\sqrt{2}},\qquada_q = \frac{\tilde{x}_q + i\,\tilde{p}_q}{\sqrt{2}}.$$In these variables, after some algebra (and up to total time derivatives), the Lagrangian takes the form:$$\begin{aligned}L &amp;= \dot{x}_c\,\tilde{p}_q - \dot{p}_c\,\tilde{x}_q \&amp;\quad -\Biggl[\delta + \frac{\chi}{2}\Bigl(x_c^2+p_c^2+\tilde{x}_q^2+\tilde{p}_q^2\Bigr)\Biggr]\,(x_c\,\tilde{x}_q+p_c\,\tilde{p}_q) \&amp;\quad + \kappa\,(x_c\,\tilde{p}_q-p_c\,\tilde{x}_q) \&amp;\quad + i\kappa\,(\tilde{x}_q^2+\tilde{p}_q^2) \&amp;\quad + 2\varepsilon\,\tilde{p}_q.\end{aligned}$$</p><hr /><h2>3. Analytic Continuation of the Quantum Fields</h2><p>To capture nonclassical (instanton) trajectories connecting metastable states, one deforms the integration contour for the quantum fields. We perform the analytic continuation$$\tilde{x}_q \to -i\,p_q,\qquad \tilde{p}_q \to i\,x_q,$$where $x_q) and $p_q) are real variables.</p><p>Let us now see how each term in the Lagrangian transforms.</p><h3>3.1 Kinetic Term</h3><p>The kinetic term$$\dot{x}_c\,\tilde{p}_q - \dot{p}_c\,\tilde{x}_q$$becomes$$\begin{aligned}\dot{x}_c\,(i\,x_q) - \dot{p}_c\,(-i\,p_q)&amp;= i\,\dot{x}_c\,x_q + i\,\dot{p}_c\,p_q\[1mm]&amp;= i\Bigl(\dot{x}_c\,x_q + \dot{p}_c\,p_q\Bigr).\end{aligned}$$</p><h3>3.2 Interaction and Nonlinear Terms</h3><p>Consider the term$$\Biggl[\delta + \frac{\chi}{2}\Bigl(x_c^2+p_c^2+\tilde{x}_q^2+\tilde{p}_q^2\Bigr)\Biggr]\,(x_c\,\tilde{x}_q+p_c\,\tilde{p}_q).$$First, note that$$\begin{aligned}x_c\,\tilde{x}_q+p_c\,\tilde{p}_q &amp;\to x_c\,(-i\,p_q)+p_c\,(i\,x_q)\[1mm]&amp;= i\,(p_c\,x_q - x_c\,p_q).\end{aligned}$$Also, since$$\tilde{x}_q^2+\tilde{p}_q^2 \to (-i\,p_q)^2+(i\,x_q)^2 = -p_q^2 - x_q^2,$$the factor in square brackets becomes$$\delta + \frac{\chi}{2}\Bigl(x_c^2+p_c^2 - (x_q^2+p_q^2)\Bigr).$$Thus, the entire interaction term transforms into$$-i\Biggl<a href="p_c\,x_q-x_c\,p_q">\delta + \frac{\chi}{2}\Bigl(x_c^2+p_c^2-x_q^2-p_q^2\Bigr)\Biggr</a>.$$</p><h3>3.3 Dissipative and Driving Terms</h3><p>The remaining terms transform as follows:</p><ul><li><p><strong>Dissipative Term:</strong>  $$  \begin{aligned}  \kappa\,(x_c\,\tilde{p}_q-p_c\,\tilde{x}_q)  &amp;\to \kappa\Bigl(x_c\,(i\,x_q)-p_c\,(-i\,p_q)\Bigr)\[1mm]  &amp;= i\,\kappa\,(x_c\,x_q+p_c\,p_q).  \end{aligned}  $$</p></li><li><p><strong>Additional Dissipative Contribution:</strong>  $$  i\kappa\,(\tilde{x}_q^2+\tilde{p}_q^2)  \to i\kappa\,\Bigl[-(x_q^2+p_q^2)\Bigr]  = -i\,\kappa\,(x_q^2+p_q^2).  $$</p></li><li><p><strong>Driving Term:</strong>  $$  2\varepsilon\,\tilde{p}_q \to 2\varepsilon\,(i\,x_q) = 2i\,\varepsilon\,x_q.  $$</p></li></ul><hr /><h2>4. The Auxiliary Hamiltonian</h2><p>Collecting the transformed terms, the Lagrangian now reads$$\begin{aligned}L \to \; &amp; i\Bigl{ \dot{x}_c\,x_q + \dot{p}_c\,p_q\[1mm]&amp;\quad - \Bigl<a href="p_c\,x_q-x_c\,p_q">\delta + \frac{\chi}{2}\bigl(x_c^2+p_c^2-x_q^2-p_q^2\bigr)\Bigr</a>\[1mm]&amp;\quad + \kappa\,(x_c\,x_q+p_c\,p_q)- \kappa\,(x_q^2+p_q^2)- 2\varepsilon\,x_q \Bigr}.\end{aligned}$$It is convenient to factor out the overall factor of $i) and to write the action in a canonical form. One introduces the notation$$\mathbf{z}_c=(x_c,p_c),\quad \mathbf{z}_q=(x_q,p_q),$$and writes the (transformed) action as$$iS = i\int dt\,L = -\int dt\Bigl[\dot{x}_c\,p_c + \dot{p}_c\,p_q - H(x_c,p_c,x_q,p_q)\Bigr].$$By comparing with the expression above, one identifies the auxiliary Hamiltonian $H(x_c,p_c,x_q,p_q)) as$$\boxed{\begin{aligned}H(x_c,p_c,x_q,p_q)=\; &amp; \Biggl(\delta + \frac{\chi}{2}\bigl(x_c^2+p_c^2-x_q^2-p_q^2\bigr)\Biggr)(p_c\,x_q - x_c\,p_q)\[1mm]&amp;\quad - \kappa\,(x_c\,x_q+p_c\,p_q)+\kappa\,(x_q^2+p_q^2)+2\varepsilon\,x_q.\end{aligned}}$$This Hamiltonian generates the saddle-point equations via$$\dot{\mathbf{z}}_c = \frac{\partial H}{\partial \mathbf{z}_q},\qquad\dot{\mathbf{z}}_q = -\frac{\partial H}{\partial \mathbf{z}_c}.$$</p><hr />
