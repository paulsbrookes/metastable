<h1>Derivation of the Keldysh Lagrangian from a Lindblad Master Equation</h1><p>This document presents a derivation of the Keldysh Lagrangian for a driven–dissipative Kerr oscillator following.</p><hr /><h2>1. Starting Point: The Lindblad Master Equation</h2><p>The Lindblad master equation for the density matrix evolution is given by</p><p>$$\partial_t \rho = -i\,[H,\rho] + \kappa\Bigl(2a\,\rho\,a^\dagger - {a^\dagger a,\rho}\Bigr),$$</p><p>where the Hamiltonian is</p><p>$$H = \delta\,a^\dagger a + \chi\,a^\dagger a^\dagger aa + i\varepsilon\,(a^\dagger - a).$$</p><p>The jump operator can be expressed as</p><p>$$L = \sqrt{2\kappa}\,a,$$</p><p>so that the dissipative part is in the standard Lindblad form.</p><hr /><h2>2. Matrix Element of the Liouvillian Superoperator</h2><p>In the coherent-state representation, we work with forward ($+$) and backward ($ - $) contours. Denoting the fields by $a_+$ and $a_-$ (with their complex conjugates $a_+^<em>$ and $a_-^</em>$), we define the matrix element of the Liouvillian superoperator as</p><p>$$L_{\rm super}(a_+^<em>,a_+,a_-^</em>,a_-) = -i\Bigl[H(a_+^<em>,a_+) - H(a_-^</em>,a_-)\Bigr] + \kappa\Bigl[2\,a_+a_-^<em> - a_+^</em>a_+ - a_-^*a_-\Bigr].$$</p><p>Here, the Hamiltonian evaluated on each contour is</p><p>$$H(a_\pm^<em>,a_\pm) = \delta\,a_\pm^</em>a_\pm + \chi\,a_\pm^{<em>2}a_\pm^2 + i\varepsilon\,(a_\pm^</em> - a_\pm).$$</p><hr /><h2>3. Constructing the Keldysh Lagrangian</h2><p>Following the recipe, the full Keldysh Lagrangian is written as</p><p>$$L_K = a_+^<em>\,i\partial_t a_+ - a_-^</em>\,i\partial_t a_- - i\,L_{\rm super}(a_+^<em>,a_+,a_-^</em>,a_-).$$</p><p>Substituting the expression for $L_{\rm super}$, we have</p><p>$$\begin{aligned}L_K &amp;= a_+^<em>\,i\partial_t a_+ - a_-^</em>\,i\partial_t a_- \&amp;\quad {} - i\Bigl{-i\Bigl[H(a_+^<em>,a_+) - H(a_-^</em>,a_-)\Bigr] + \kappa\Bigl[2\,a_+a_-^<em> - a_+^</em>a_+ - a_-^<em>a_-\Bigr]\Bigr} \&amp;= a_+^</em>\,i\partial_t a_+ - a_-^<em>\,i\partial_t a_- + \Bigl[H(a_+^</em>,a_+) - H(a_-^<em>,a_-)\Bigr] \&amp;\quad {} - i\,\kappa\Bigl[2\,a_+a_-^</em> - a_+^<em>a_+ - a_-^</em>a_-\Bigr].\end{aligned}$$</p><p>Inserting the explicit form of the Hamiltonian difference</p><p>$$\begin{aligned}H(a_+^<em>,a_+) - H(a_-^</em>,a_-) &amp;= \delta\,\Bigl(a_+^<em>a_+ - a_-^</em>a_-\Bigr) + \chi\,\Bigl(a_+^{<em>2}a_+^2 - a_-^{</em>2}a_-^2\Bigr) \&amp;\quad {} + i\varepsilon\Bigl[(a_+^<em> - a_+) - (a_-^</em> - a_-)\Bigr],\end{aligned}$$</p><p>the final form of the Keldysh Lagrangian becomes</p><p>$$\boxed{\begin{aligned}L_K =\; &amp; a_+^<em>\,i\partial_t a_+ - a_-^</em>\,i\partial_t a_- \&amp;\quad {} - \delta\,\Bigl(a_+^<em>a_+ - a_-^</em>a_-\Bigr) - \chi\,\Bigl(a_+^{<em>2}a_+^2 - a_-^{</em>2}a_-^2\Bigr) \&amp;\quad {} - i\varepsilon\,\Bigl[(a_+^<em> - a_+) - (a_-^</em> - a_-)\Bigr] \&amp;\quad {} - i\,\kappa\Bigl[2\,a_+a_-^<em> - a_+^</em>a_+ - a_-^*a_-\Bigr].\end{aligned}}$$</p><p>This expression clearly separates:- The <strong>time-evolution</strong> terms: $a_+^<em>\,i\partial_t a_+ - a_-^</em>\,i\partial_t a_-$,- The <strong>coherent Hamiltonian</strong> contributions, and- The <strong>dissipative</strong> contributions.</p><hr />
