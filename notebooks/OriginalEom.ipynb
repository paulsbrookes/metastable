<h1>Derivation of the Equations of Motion</h1><p>In this document we explain how to derive the equations of motion from the Lagrangian</p><p>$$\begin{aligned}L &amp;= \dot{x}_c\,\tilde{p}_q - \dot{p}_c\,\tilde{x}_q - \Biggl{\delta  + \frac{\chi}{2} \Bigl(x_c^2 + p_c^2 + \tilde{x}_q^2 + \tilde{p}_q^2 \Bigr) \Biggr}(x_c\,\tilde{x}_q + p_c\,\tilde{p}_q) \&amp;\quad + \kappa (x_c\,\tilde{p}_q - p_c\,\tilde{x}_q)  + i \kappa (\tilde{x}_q^2 + \tilde{p}_q^2) + 2 \varepsilon\,\tilde{p}_q\,.\end{aligned}$$</p><h2>1. Recognizing the Canonical Structure</h2><p>Notice that the Lagrangian is <em>first order</em> in time derivatives, with the kinetic (or "symplectic") part given by</p><p>$$L_{\rm kin} = \dot{x}_c\,\tilde{p}_q - \dot{p}_c\,\tilde{x}_q\,.$$</p><p>This form identifies the canonical pairs as:- $x_c$ with conjugate momentum $\tilde{p}_q$, and- $p_c$ with conjugate momentum $-\tilde{x}_q$.</p><p>Thus, the Lagrangian can be rewritten as</p><p>$$L = \tilde{p}_q\,\dot{x}_c - \tilde{x}_q\,\dot{p}_c - H(x_c, p_c, \tilde{x}_q, \tilde{p}_q)\,,$$</p><p>where the "Hamiltonian" is given by</p><p>$$\begin{aligned}H(x_c, p_c, \tilde{x}_q, \tilde{p}_q) = &amp;\; \Biggl<a href="x_c\,\tilde{x}_q+p_c\,\tilde{p}_q">\delta + \frac{\chi}{2} \Bigl(x_c^2+p_c^2+\tilde{x}_q^2+\tilde{p}_q^2\Bigr)\Biggr</a> \&amp;\; -\kappa(x_c\,\tilde{p}_q-p_c\,\tilde{x}_q) - i\kappa(\tilde{x}_q^2+\tilde{p}_q^2) - 2\varepsilon\,\tilde{p}_q\,.\end{aligned}$$</p><h2>2. Writing Hamilton's Equations</h2><p>Since the kinetic term is in canonical form, the Euler–Lagrange equations for our variables are equivalent to Hamilton's equations. In particular, we have</p><p>$$\begin{aligned}\dot{x}_c &amp;= \frac{\partial H}{\partial \tilde{p}_q}\,, &amp;\quad \dot{\tilde{p}}_q &amp;= -\frac{\partial H}{\partial x_c}\,,\[1mm]\dot{p}_c &amp;= -\frac{\partial H}{\partial \tilde{x}_q}\,, &amp;\quad \dot{\tilde{x}}_q &amp;= \frac{\partial H}{\partial p_c}\,.\end{aligned}$$</p><h2>3. Computing the Partial Derivatives</h2><p>To compute these derivatives efficiently, one may introduce the shorthand</p><p>$$A \equiv \delta+\frac{\chi}{2}\Bigl(x_c^2+p_c^2+\tilde{x}_q^2+\tilde{p}_q^2\Bigr)\quad \text{and} \quadB \equiv x_c\,\tilde{x}_q+p_c\,\tilde{p}_q\,.$$</p><h3>Derivative with Respect to $\tilde{p}_q$:</h3><p>$$\frac{\partial H}{\partial \tilde{p}_q}= \chi\,\tilde{p}_q\,B + A\,p_c - \kappa\,x_c - 2i\kappa\,\tilde{p}_q - 2\varepsilon\,.$$</p><p>Thus, the equation of motion for $x_c$ is</p><p>$$\dot{x}_c = \chi\,\tilde{p}_q\,B + A\,p_c - \kappa\,x_c - 2i\kappa\,\tilde{p}_q - 2\varepsilon\,.$$</p><h3>Derivative with Respect to $x_c$:</h3><p>$$\frac{\partial H}{\partial x_c}= \chi\,x_c\,B + A\,\tilde{x}_q - \kappa\,\tilde{p}_q\,.$$</p><p>Thus, the equation of motion for $\tilde{p}_q$ is</p><p>$$\dot{\tilde{p}}_q = -\chi\,x_c\,B - A\,\tilde{x}_q + \kappa\,\tilde{p}_q\,.$$</p><h3>Derivative with Respect to $\tilde{x}_q$:</h3><p>$$\frac{\partial H}{\partial \tilde{x}_q}= \chi\,\tilde{x}_q\,B + A\,x_c + \kappa\,p_c - 2i\kappa\,\tilde{x}_q\,.$$</p><p>Thus, the equation of motion for $p_c$ is</p><p>$$\dot{p}_c = -\chi\,\tilde{x}_q\,B - A\,x_c - \kappa\,p_c + 2i\kappa\,\tilde{x}_q\,.$$</p><h3>Derivative with Respect to $p_c$:</h3><p>$$\frac{\partial H}{\partial p_c}= \chi\,p_c\,B + A\,\tilde{p}_q + \kappa\,\tilde{x}_q\,.$$</p><p>Thus, the equation of motion for $\tilde{x}_q$ is</p><p>$$\dot{\tilde{x}}_q = \chi\,p_c\,B + A\,\tilde{p}_q + \kappa\,\tilde{x}_q\,.$$</p><h2>4. Final Set of Equations</h2><p>Summarizing, the equations of motion are</p><p>$$\boxed{\begin{aligned}\dot{x}_c &amp;= \chi\,\tilde{p}_q\,(x_c\,\tilde{x}_q+p_c\,\tilde{p}_q)+\left(\delta+\frac{\chi}{2}(x_c^2+p_c^2+\tilde{x}_q^2+\tilde{p}_q^2)\right)p_c-\kappa\,x_c-2i\kappa\,\tilde{p}_q-2\varepsilon\,,\[1mm]\dot{\tilde{p}}_q &amp;= -\chi\,x_c\,(x_c\,\tilde{x}_q+p_c\,\tilde{p}_q)-\left(\delta+\frac{\chi}{2}(x_c^2+p_c^2+\tilde{x}_q^2+\tilde{p}_q^2)\right)\tilde{x}_q+\kappa\,\tilde{p}_q\,,\[1mm]\dot{p}_c &amp;= -\chi\,\tilde{x}_q\,(x_c\,\tilde{x}_q+p_c\,\tilde{p}_q)-\left(\delta+\frac{\chi}{2}(x_c^2+p_c^2+\tilde{x}_q^2+\tilde{p}_q^2)\right)x_c-\kappa\,p_c+2i\kappa\,\tilde{x}_q\,,\[1mm]\dot{\tilde{x}}_q &amp;= \chi\,p_c\,(x_c\,\tilde{x}_q+p_c\,\tilde{p}_q)+\left(\delta+\frac{\chi}{2}(x_c^2+p_c^2+\tilde{x}_q^2+\tilde{p}_q^2)\right)\tilde{p}_q+\kappa\,\tilde{x}_q\,.\end{aligned}}$$</p>
